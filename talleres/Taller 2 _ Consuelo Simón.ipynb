PREGUNTA 2.3: Realiza un test de hipotesis para evaluar si las dos variables que se agregaron son
significativas para el modelo simultaneamente. Use significancia del 5%.

```r
F_critico <- qf (0.95 , 2 , 203)
R2_r <- 0.03191
R2_ur <- 0.04059
F_est <- ( R2_ur - R2_r )/(1 - R2_ur )* 203 /2
F_est
F_critico
if ( F_est > F_critico ) {"Se rechaza H0"} else {"No se rechaza H0"}
```

Como F_est es menor que F_critico, no se rechaza la hipótesis nula. Esto indica que las variables agregadas (sales_mayor_promedio y su interacción con roe) no son conjuntamente significativas al 5%, por lo que no mejoran el modelo de forma relevante.

PREGUNTA 2.4:
Volviendo a la regresión del salario en función de sales, roe y ros. Considere
dos grupos, Grupo 1: Aquellos clientes que sales es mayor al promedio, y Grupo 2: Aquellos
clientes que sales es menor al promedio. Realice un test de hipotesis para grupos al 5% de
significancia. Use el metodo que más le acomode.

```r
ceosal1$sales_D <- ceosal1$sales * ceosal1$sales_mayor_promedio
ceosal1$roe_D <- ceosal1$roe * ceosal1$sales_mayor_promedio
ceosal1$ros_D <- ceosal1$ros * ceosal1$sales_mayor_promedio

modelo <- lm(
  salary ~ sales + roe + ros +
    sales_mayor_promedio +
    sales_D + roe_D + ros_D,
  data = ceosal1
)

summary(modelo)

F_critico <- qf(0.95, 4, 201)

R2_ur <- 0.06028
R2_r <- 0.03191

F_est <- ((R2_ur - R2_r) / (1 - R2_ur)) * (201 / 4)

F_est
F_critico
if ( F_est > F_critico ) {"Se rechaza H0"} else {"No se rechaza H0"}
```

Como F_est es menor que F_critico, no se rechaza H0 con un 5% de significancia. Esto indica que no hay evidencia suficiente para afirmar que el comportamiento del modelo es distinto entre empresas con ventas altas y bajas.

PREGUNTA 2.5:
Usando la regresión del salario en función de sales, roe y ros, se distinguen los
grupos mediante las dummies; indus, finance, consprod y utility, que corresponden al tipo de
empresa. Realice mediante test de Chow, con 5% de significancia, si los modelos para cada
grupo son iguales.

```r
ceosal1_indus <- subset(ceosal1, indus == 1)
ceosal1_finance <- subset(ceosal1, finance == 1)
ceosal1_consprod <- subset(ceosal1, consprod == 1)
ceosal1_utility <- subset(ceosal1, utility == 1)

modelo_r <- lm(salary ~ sales + roe + ros, data = ceosal1)

modelo_indus <- lm(salary ~ sales + roe + ros, data = ceosal1_indus)
modelo_finance <- lm(salary ~ sales + roe + ros, data = ceosal1_finance)
modelo_consprod <- lm(salary ~ sales + roe + ros, data = ceosal1_consprod)
modelo_utility <- lm(salary ~ sales + roe + ros, data = ceosal1_utility)

SSR_r <- sum(residuals(modelo_r)^2)

SSR_indus <- sum(residuals(modelo_indus)^2)
SSR_finance <- sum(residuals(modelo_finance)^2)
SSR_consprod <- sum(residuals(modelo_consprod)^2)
SSR_utility <- sum(residuals(modelo_utility)^2)

SSR_ur <- SSR_indus + SSR_finance + SSR_consprod + SSR_utility

G <- 4
k <- 3
n <- nrow(ceosal1)

gl1 <- (G - 1) * (k + 1)
gl2 <- n - G * (k + 1)

F_est <- ((SSR_r - SSR_ur) / gl1) / (SSR_ur / gl2)
F_critico <- qf(0.95, gl1, gl2)

F_est
F_critico
if ( F_est > F_critico ) {"Se rechaza H0"} else {"No se rechaza H0"}
```

Con el test de chow realizado podemos afirmar que no se rechaza H0 con un 5% de significancia, por lo que no hay evidencia suficiente para afirmar que los coeficientes del modelo difieren entre los distintos tipos de empresas. Esto sugiere que la relación entre salario, ventas y rentabilidad es similar entre los sectores analizados.
